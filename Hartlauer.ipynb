### Die Daten werden von der Website des ARBÖ geladen.

```python
url = "https://www.hartlauer.at/produkt/sony-alpha-6000-selp-16-50-3-5-5-6-oss-pz-schwarz/ERP_ARTICLE_190068930.html"
```

Wir geben im Skript vor ein Google Chrome Browser zu sein.

```python
from fake_useragent import UserAgent
ua = UserAgent()
agent = ua.chrome
print(agent)
headers = {'User-Agent': str(agent)}
```

### Betrachten wir die Seite

Sie enthält JavaScript Code der die Daten enthält.

```python
page = requests.get(url, headers = headers)
print(page.text)
```

Wir parsen den HTML Code. Und suchen den Tag mit der Id der das Java Script entält.

```python
soup = BeautifulSoup(page.content, "html.parser")
results = soup.find_all("h1", {"class": "c-mainProductInfo__title product-name ias-productName"});
results = str(results[0]).split('>')
results = str(results[1]).split('<')
print("Produktname:", results[0])

results1 = soup.find_all('span',{"class": "c-priceContainer__price ias-productPrice"})
results1 = str(results1[0]).split('>')
results1 = str(results1[1]).split('<')
print("Preis:", results1[0])
```
